```python
def convolve2d( image, kernel):
    flippedK = np.flipud(np.fliplr(kernel))
    padsize = int(kernel.shape[0]/2)
    img_convoluted = np.zeros_like(image)
    img_padded = np.zeros((image.shape[0]+2*padsize, image.shape[1]+2*padsize))
    img_padded[1:-1, 1:-1] = image
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            img_convoluted[i,j] = (flippedK*img_padded[i:i+kernel.shape[0],j:j+kernel.shape[1]]).sum()
    return img_convoluted
```
